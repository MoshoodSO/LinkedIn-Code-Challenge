```python
# Python code below
# Use print("messages...") to debug your solution.

show_expected_result = False
show_hints = False

import calendar

def count_days(year, month, whichday):
    '''
    Return number of instances of that day od the week in a given month of a given year

    Args:
        year (int)     : The given year
        month (int)    : The given month of the year
        whichday (int) : The day of the week

    Returns:
        int: The number of that day of the week

    Examples:
    >>> count_days(2025,12,0)
    >>> 5

    >>> count_days(2030,1,5)
    >>> 4
    '''
    counter = 0
    cal = calendar.monthcalendar(year, month)
    for i in range(len(cal)):
        if cal[i][whichday] != 0:
            counter += 1
    return counter
```
