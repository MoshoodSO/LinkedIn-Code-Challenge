```python
def factorial(num):
    '''
    Calculate the factorial of positive whole number 

    Args:
        num (int): A positive whole number

    Returns:
        int: A whole number.

    Examples:
    >>> factorial(6)
    >>> 720

    >>> factorial("span span span")
    >>> None
    '''
    if isinstance(num, int) == True:
        if num == 1 or num == 0:
            return 1
        elif num > 0:
            return factorial(num-1) * num
    else:
        pass
```
