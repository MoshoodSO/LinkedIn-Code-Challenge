```python
# Python code below
# Use print("messages...") to debug your solution.

show_expected_result = False
show_hints = False

def find_largest(test_strings):
    '''
    It returns the length of the longest string in a list

    Args:
        test_strings (list): A list of strings

    Returns:
        int: A whole number (The length of the longest string)

    Example:
    test_strings = [
    "Hello World!",
    "And how can this be? For he is the Kwisatz Haderach!",
    "Four score and seven years ago",
    "Life moves pretty fast. If you dont stop and look around once in a while, you could miss it."
    ]
    >>> find_largest(test_strings)
    >>> 93
    '''
    max_len = 0
    for word in test_strings:
        if len(word) > max_len:
            max_len = len(word)
    return max_len
```

```python
test_strings = [
    "Hello World!",
    "And how can this be? For he is the Kwisatz Haderach!",
    "Four score and seven years ago",
    "Life moves pretty fast. If you dont stop and look around once in a while, you could miss it."
]
```
