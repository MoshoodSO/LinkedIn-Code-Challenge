```python
# Python code below
# Use print("messages...") to debug your solution.

show_expected_result = False
show_hints = False

def count_numbers(which, numbers):
    '''
    Count the number of even or odd numbers in a string, based upon the string argument

    Args:
        which (str) :
        number (list): A al positive whole number

    Returns:
        int: A whole number.

    Examples:
    >>> factorial(6)
    >>> 720

    >>> factorial("span span span")
    >>> None
    '''
    even_counter = 0
    odd_counter = 0
    if which not in ["even", "odd"]:
        return -1
    for num in numbers:
        if num % 2 == 0:
            even_counter += 1
        else: 
            odd_counter += 1
    if which == "even":
        return even_counter
    return odd_counter
```

```python
count_numbers("even", [2,5,20,30,55])
```

```python
count_numbers("blarg", [20,30,55])
```

```python
count_numbers("odd", [2,5,20,30,55])
```
