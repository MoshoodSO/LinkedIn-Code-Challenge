```python
text = "AABBBAAABC"
```

```python
stringVal = 'AAAAABBBBBAAACCCCC'
```

```python
def encodeString(stringVal):
    lst=[]
    prevChar = None
    for i, char in enumerate(stringVal):
        if char != prevChar:
            lst.append([char, i])
        prevChar = char

    lst = lst + [[" ", len(stringVal)]]
    return [(lst[i][0], lst[i+1][1] - lst[i][1]) for i in range(len(lst)-1)]
```

```python
encodeString(stringVal)
```

```python
encodeString(text)
```

```python
def decodeString(lst):
    listVal = [i[0]*i[1] for i in lst]
    return "".join(listVal)
```

```python
decodeString([('A', 5), ('B', 5), ('A', 3), ('C', 5)])
```

```python
decodeString([('A', 2), ('B', 3), ('A', 3), ('B', 1), ('C', 1)])
```
