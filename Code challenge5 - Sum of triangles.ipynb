```python
# Python code below
def triangle(num):
    """
    A function that returns triangular numbers (sum of the number less than or equals to it).

    Args:
        num (int): a whole number greater than 0

    Returns:
        int: A number.

    Examples:
    >>> triangle(3)
    >>> 6

    >>> triangle(8)
    >>> 36

    """
    if num == 1:
        return num
    return num + triangle(num - 1)

    
```

```python
def square(num):
    """
    A function that returns the square of a number using triange function above.

    Args:
        num (int): a whole number greater than 0

    Returns:
        int: A number.

    Examples:
    >>> square(3)
    >>> 9

    >>> square(8)
    >>> 64

    """
    return triangle(num) + triangle(num-1)
```
