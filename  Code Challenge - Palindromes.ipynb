```python
# Python code below
# Use print("messages...") to debug your solution.

show_expected_result = False
show_hints = False

def is_palindrome(teststr):
    '''
    Returns a Boolean True or False value, indicating whether a string is a palindrome or not

    Args:
        teststr (str): A string (word irrespective of case, punctuation marks or spaces)

    Returns:
        Bool: True/False.

    Examples:
    >>> is_palindrome("Radar!")
    >>> True

    >>> test_word = "Madam, I'm Adam."
    >>> is_palindrome(test_word)
    >>> True
    '''
    
    clean_str = teststr.lower()
    result = ''
    for i in clean_str:
        if i in string.ascii_lowercase:
            result+=i
    #if result != clean_str:
     #   return "The input is an alphanumeric character!"
    return result == result[::-1]
```

```python
is_palindrome("Radar!")
```
